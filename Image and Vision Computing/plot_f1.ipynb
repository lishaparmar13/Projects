```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pickle
import numpy as np
```

```python
class_names = ['black widow', 'captain america', 'doctor strange', 'hulk', 'ironman', 'loki', 'spider-man', 'thanos']
```

```python
y = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
```

```python
with open('svm_f1\occulsion_f1.pkl', 'rb') as f:
    loaded_dict = pickle.load(f)
```

```python
arr = []
```

```python
for i in range(len(class_names)):

    arr.append([elem[i] for elem in loaded_dict.values()])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

for i in range(len(class_names)):
    ax.plot(y,arr[i])

plt.title(" Pertubations on SVM + SIFT")
ax.legend(['black widow', 'captain america', 'doctor strange', 'hulk', 'ironman', 'loki', 'spider-man', 'thanos'],loc = 'upper right')
plt.xlabel("Occulsion")
plt.ylabel("F1-Score")    
plt.show()
```
